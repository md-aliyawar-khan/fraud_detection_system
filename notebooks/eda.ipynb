```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r'C:\Users\aliya\Desktop\fraud_detection_project\data\raw\creditcard_2023.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
display(data.head())

# Display basic information about the dataset
print("\nDataset Information:")
print(data.info())

# Display basic statistics of the dataset
print("\nDataset Statistics:")
display(data.describe())

# Check for missing values
print("\nMissing Values in the Dataset:")
print(data.isnull().sum())
```

```python
# Distribution of target variable 'Class'
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=data)
plt.title('Distribution of Fraud vs Non-Fraud Transactions')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
# Check for imbalance in the dataset
fraud_percentage = data['Class'].value_counts(normalize=True) * 100
print(f"\nPercentage of Fraud vs Non-Fraud Transactions:\n{fraud_percentage}")

```

```python
# Plot the distribution of the 'Amount' feature
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

```python


# Plot the correlation heatmap between features
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Features')
plt.show()


```

```python
# Pairplot to explore feature distributions and relationships for a subset of features
subset = data[['V1', 'V2', 'V3', 'V4', 'V5', 'Class']]  # Using first few features for visualization
sns.pairplot(subset, hue='Class', plot_kws={'alpha':0.5})
plt.title('Pairplot of Selected Features')
plt.show()
```

```python
# Display skewness of numerical features (useful for potential feature engineering)
skewness = data.skew().sort_values(ascending=False)
print("\nSkewness of Numerical Features:")
print(skewness)
```

```python
# Correlation of 'Amount' with 'Class'
plt.figure(figsize=(6, 4))
sns.boxplot(x='Class', y='Amount', data=data)
plt.title('Transaction Amount vs Class')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Transaction Amount')
plt.show()
```
